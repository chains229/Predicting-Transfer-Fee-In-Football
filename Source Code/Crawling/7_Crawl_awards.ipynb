```python
def get_soup(URL):
    
    chrome_options = Options()
    chrome_options.add_argument('--window-size=1,1')
    
    #options = webdriver.ChromeOptions()
    #options.add_argument('--disable-popup-blocking')
    
    driver = webdriver.Chrome(ChromeDriverManager().install(), options = chrome_options)
    wait = WebDriverWait(driver, 10)
    driver.get(URL)
    if len(driver.find_elements_by_css_selector("[name=ROBOTS]")) > 0:
        driver.get(URL)
    soup = BeautifulSoup(driver.page_source)
    driver.close()
    return soup
```

```python
# Hàm thu thập tổng số danh hiệu cá nhân và tập thể của từng cầu thủ
def get_award(players_df):
    columns = ['player', 'award']
    
    award_df = pd.DataFrame(columns = columns)

    for index in range(players_df.shape[0]):
        each_link = players_df.iloc[index]['link']
        
        new_link = each_link.replace("profil","erfolge")
        new_link = 'https://www.transfermarkt.co.uk' + new_link
        print(new_link)
        
        soup = get_soup(new_link)
        
        tables = soup.find_all('table', class_ = 'auflistung')
        awards = 0
        for table in tables:
            body = table.find('tbody')
            award = len(body.find_all('tr')) 
            awards += award
        
        data = {'player': new_link, 'award': awards}
        
        award_each = pd.Series(data)
        
        award_df = award_df.append(award_each, ignore_index=True)
    
    return award_df
```

```python
# Sample_data là một list, mỗi phần tử là 1 dataframe tách ra từ sampled_players
sample_data = []
sample_data.append(sampled_players.iloc[:10])
i = 10
while i < sampled_players.shape[0]:
    if i >= sampled_players.shape[0] - 200:
        sample_data.append(sampled_players.iloc[i:])
    else:
        sample_data.append(sampled_players.iloc[i:i+200])
    i += 200 # mỗi sample chứa 200 dòng
print(len(sample_data))
```

```python
i = 40
getting_awards = get_award(sample_data[i])
getting_awards
```
