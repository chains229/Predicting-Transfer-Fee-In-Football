```python
import pandas as pd
import numpy as np
```

```python
players = pd.read_csv("C:/Users/Dell/Downloads/Đồ án/Data/Full/players_transfermarket.csv")
basic_info =  pd.read_csv("C:/Users/Dell/Downloads/Đồ án/Data/Full/basic_info_with_link.csv")
stats =  pd.read_csv("C:/Users/Dell/Downloads/Đồ án/Data/Full/player_stats.csv")
injuries = pd.read_csv("C:/Users/Dell/Downloads/Đồ án/Data/Full/Injuries.csv")
awards = pd.read_csv("C:/Users/Dell/Downloads/Đồ án/Data/Full/awards.csv")
market_values = pd.read_csv("C:/Users/Dell/Downloads/Đồ án/Data/Full/market_value.csv")
```

Khai báo những file đã được thu thập trước đó và loại bỏ những cột thừa

```python
basic_info = basic_info.drop(basic_info.columns[[0, 1, 3, 5]], axis=1)
basic_info.rename(columns={'link': 'player'}, inplace=True)
basic_info
```

```python
stats = stats.drop(stats.columns[[0]], axis=1)
stats
```

```python
injuries = injuries.drop(injuries.columns[[0]], axis=1)
injuries
```

```python
market_values = market_values.drop(market_values.columns[[0]], axis=1)
market_values.rename(columns={'link': 'player'}, inplace=True)
market_values
```

```python
awards['player'] = awards['player'].str.replace('https://www.transfermarkt.co.uk', '')
awards['player'] = awards['player'].str.replace('erfolge', 'profil')
awards = awards.drop(awards.columns[[0]], axis=1)
awards
```

Gộp các DataFrame lại thành 1 

```python
combined_df = pd.merge(basic_info, stats, on='player')
combined_df = pd.merge(combined_df, injuries, on='player')
combined_df = pd.merge(combined_df, awards, on='player')
combined_df = pd.merge(combined_df, market_values, on='player')
combined_df
```

Xử lý các giá trị null

```python
# Get the columns with null values
columns_with_null = combined_df.columns[combined_df.isnull().any()].tolist()

# Print the columns with null values
print(columns_with_null)
```

```python
# Tiền xử lý cột position
def position(s):
    if s == 'Goalkeeper':
        return 1
    elif 'Defender' in s:
        return 2
    elif 'Midfield' in s:
        return 3
    else:
        return 4
def check_winger(s):
    if 'Left' in s or 'Right' in s:
        return 1
    else:
        return 0

combined_df['position_encoded'] = combined_df['position'].apply(position)
combined_df['winger'] = combined_df['position'].apply(check_winger)
```

```python
# Tiền xử lý cột height, age
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy = 'mean')
columns_with_missing = combined_df.columns[combined_df.isnull().any()].tolist()
combined_df[columns_with_missing] = imputer.fit_transform(combined_df[columns_with_missing])
```
